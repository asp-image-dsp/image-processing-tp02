## Cuantización del **cráneo**

```python
def quantize(image, from_levels, to_levels):
    """ Converts the amount of color levels used to represent an image, modifying its quantization levels.
        @param image Image to be converted
        @param from_levels Current amount of color levels
        @param to_levels New amount of color levels
        @return New image
    """
    to_levels = to_levels - 1
    from_levels = from_levels - 1
    return np.round(image * (to_levels / from_levels)) * (from_levels / to_levels)
```
