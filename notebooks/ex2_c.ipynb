```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
f = np.zeros((30,30))
f[5:24, 13:17] = 1
```

```python
plt.imshow(f, cmap='Greys')
```

```python
F = np.fft.fft2(f, [256, 256])      #Apply 2 dimensional FFT
F = np.fft.fftshift(F)  #Center FFT
print(np.min(np.abs(F)))
print(np.max(np.abs(F)))
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
pos = ax.imshow(np.abs(F), cmap=plt.cm.jet)
fig.colorbar(pos, ax=ax, shrink=0.8)
plt.axis('off')
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
pos = ax.imshow(np.log(1+np.abs(F)), cmap=plt.cm.jet)
fig.colorbar(pos, ax=ax, shrink=0.8)
plt.
plt.axis('off')
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
pos = ax.imshow(np.angle(F), cmap=plt.cm.jet)
fig.colorbar(pos, ax=ax, shrink=0.8)
plt.axis('off')
```
